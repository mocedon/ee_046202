#### <img src="https://img.icons8.com/dusk/64/000000/idea.png" style="height:30px;display:inline"> Solution 1 - Hypothesis Testing
---
1. 
* Denote $ \underline{x} := \{x_i\}_{i=1}^n $ and $ \underline{X} := \{X_i\}_{i=1}^n $
* Recall from the lecture:
$$ \text{p-value} \triangleq \inf_{T(\underline{x}) \geq c} \left\{ \alpha(c) \right\} $$
Therefore, there exist some $ \tilde{c} $ that satisfies $ T(\underline{x}) \geq \tilde{c} $ for which $ \text{p-value} = \alpha(\tilde{c}) $
* Additionally recall 
$$ \alpha(c) \triangleq \sup_{\theta \in \Theta_0} P_{\theta}\left( T(\underline{X}) \geq c) \right) $$
* Since $ \alpha(c) $ is a monotonically decreasing function of $ c $, the bigger the $ c $ we take, the smaller the $ \alpha $ we get. But we can't take $ c $ any bigger than $ T(\underline{x}) $ or the condition $ T(\underline{x}) \geq c $ won't hold, thus $ \tilde{c} := T(\underline{x}) $ is the maximal $ c $ available, and it is providing the minimal $ \alpha $. And now:
$$ \text{p-value} = \alpha(\tilde{c}) = \alpha(T(\underline{x})) = \sup_{\theta \in \Theta_0} P_{\theta}\left( T(\underline{X}) \geq  T(\underline{x}) \right) $$

2. 
**Pearson's Chi-Squared Test approach**
* Here we use a Pearson's Chi-Squared test with 5 degrees of freedom.
    * Pearson's Chi-Squared test because the data is categorical, it is the frequency of occurrences of non-ordered data, i.e. the value of the rolled number is meaningless for this matter.
    * 6 categories for 6 possible roll outcomes.
    * 5 degrees of freedom, as they are the number of categories minus 1.
* Null & Alternative Hypotheses:
\begin{align}
&H_0: \text{Die unbiased = each category has a likelihood of occurrence of } \frac{1}{6}. \\
&H_1: \text{Die biased = any other PMF}.
\end{align}
* The Chi-Squared Statistic: $$ \chi^2_{stat} = \sum_{i=1}^n \frac{(O[i] - E[i])^2}{E[i]} $$
    * $O[i]$ - Observed data ($i^{th}$ category).
    * $E[i]$ - Expected data (for the $i^{th}$ category).
* Our data:

|         |1 |2 |3 |4 |5 |6 |Total|
|---------|--|--|--|--|--|--|-----|
|Observed |8 |4 |11|9 |11|17|60   |
|Expected |10|10|10|10|10|10|60   |

* Calculating Chi-Squared statistic value and p-value:
\begin{align}
\chi^2_{stat} &= \frac{(8-10)^2}{10} + \frac{(4-10)^2}{10} + \frac{(11-10)^2}{10} + \frac{(9-10)^2}{10} + \frac{(11-10)^2}{10} + \frac{(17-10)^2}{10} \\
&= 0.4 + 3.6 + 0.1 + 0.1 + 0.1 + 4.9 \\
&= 9.2 \\
\chi^2_{5, 1-\text{p-value}} &\stackrel{!}{=} 9.2 \\
\text{p-value} &= P(\chi^2_5 \geq 9.2) \\
&\approx 0.1
\end{align}
* We should find critical Chi-Squared values for **confidence levels** $(1-\alpha) $ of $ 99\%, 95\% $, which correspond to significance levels $ (\alpha) $ of $ 0.01, 0.05 $. Using the provided Chi-Squared table:
    * $ \chi^2_{5, 1-0.01} \stackrel{table}{=} 15.09 $
    * $ \chi^2_{5, 1-0.05} \stackrel{table}{=} 11.07 $
    * Since $ \chi^2_{stat} = 9.2 $ is smaller than both, in both cases we don't have enough evidence to reject $H_0$ and by that to accuse the die to be unbiased. And this is no surprise since both significance levels are lower than the p-value.

**Log-Likelihood approach**:
* Assume a general 6-sided die: $ X \sim (p_1,p_2,p_3,p_4,p_5,p_6) $ <br> where $ \theta = (p_1,p_2,p_3,p_4,p_5,p_6) $ is the parameter vector.<br>
This is a random variable with the following PMF:
\begin{cases} 
\forall i \in \{1,...,6\}: P(X=i) = p_i \\ 
\sum_{i=1}^6 p_i = 1 \\
\end{cases}
* Our random variable is the counter of outcomes of $ N = 60 $ rolls:
$$ Y := \Big(\sum_{j=1}^N \mathbb{1}_{\{X_j=1\}}\Big)_{i=1}^6 \sim Multi(N,p_1,...,p_6) $$
$$
{{P}_{Y}}({{k}_{1}},..,{{k}_{6}})=\frac{N!}{{{k}_{1}}!\cdots !{{k}_{6}}!}{{p}_{1}}^{{{k}_{1}}}\cdots {{p}_{6}}^{{{k}_{6}}}\text{   When }\sum\nolimits_{i=1}^{6}{{{k}_{i}}}=N
$$
* Given a realization $ y = (k_1,...k_6) $, the Likelihood function becomes:
$$ L(y|\theta) = L(y|{{p}_{1}},..,{{p}_{6}})=\frac{N!}{{{k}_{1}}!\cdots !{{k}_{6}}!}{{p}_{1}}^{{{k}_{1}}}\cdots {{p}_{6}}^{{{k}_{6}}} $$
* Since $ H_0 $ is a simple hypothesis  ($ \theta_0 = (\frac{1}{6},...\frac{1}{6}) $ -> a single value!), it's Maximum Likelihood Estimator is simply $$ \hat{\theta}_0 = (\frac{1}{6},...\frac{1}{6}) $$
However $ H_1 $ a composite hypothesis. Since and $ \Theta_0 \cup \Theta_1 $ has no constraints besides the multinomial usual ones, the general Maximum Likelihood Estimator becomes: $$ \hat{\theta} = (\frac{k_1}{N},...,\frac{k_6}{N}) $$
* The Likelihood-Ratio Test Statistic function becomes:
$$ \Lambda (y) \triangleq \frac{L(y|{{{\hat{\theta }}}_{0}})}{L(y|\hat{\theta })}={{\left( \frac{N}{6k_{1}} \right)}^{{{k}_{1}}}}\cdots {{\left( \frac{N}{6k_{6}} \right)}^{{{k}_{6}}}} $$
* Thanks to the fact that $H_0$ is a simple hypothesis, the Wilks Theorem sustains that for a sufficiently large $N$, the distribution of $\lambda(y) \triangleq -2log \Lambda (y) $ converges into the distribution of $ \chi^2_5 $, DOF=5 as the number of degrees of freedom of $\theta$:
\begin{cases}
+6 \text{ for } (p_1,\cdots,p_6) \\
-1 \text{ for the constaint of a probability vector: } \sum_{i=1}^6 p_i = 1
\end{cases}
Expressing $\lambda(y)$:
\begin{align}
\lambda (y) &= -2\log \Lambda (y) \\
&= \sum\limits_{i=1}^{6}{2{{k}_{i}}\log (\frac{6{{k}_{i}}}{N})} \sim \chi^2_5
\end{align}
* Now the test of size $\alpha$ is this: $$ \text{reject }H_0 \text{ if } \lambda (y) \geq \chi^2_{5,\alpha} $$
* Calculating statistic value and p-value:
\begin{align}
\lambda (y) &= 2(8log0.8 + 4log0.4 + 11log1.1 + 9log0.9 + 11log1.1 + 17log1.7) \\
&= 9.44
\\
\chi^2_{5, 1-\text{p-value}} &\stackrel{!}{=} 9.44 \\
\text{p-value} &= P(\chi^2_5 \geq 9.44) \\
&\approx 0.1
\end{align}

* Critical values remained the same:
    * $ \chi^2_{5, 1-0.01} \stackrel{table}{=} 15.09 $
    * $ \chi^2_{5, 1-0.05} \stackrel{table}{=} 11.07 $
    * Since $ \chi^2_{stat} = 9.44 $ is smaller than both, in both cases we don't have enough evidence to reject $H_0$ and by that to accuse the die to be unbiased. And this is no surprise since both significance levels are lower than the p-value. **This is a further no surprise**, because we expected the Chi-Squared-Test and the Log-Likelihood-Test to be consistent!

\chi^2_{5, 1-\text{p-value}} &\stackrel{!}{=} 92 \\

* The Chi-Squared critical values are the same as before, since both significance levels and degrees of freedom have remained the same.

    * $ \chi^2_{5, 1-0.05} \stackrel{table}{=} 11.07 $

**Log-Likelihood approach**:
* Here the x10 multiplication only affects the powers in the likelihood functions and the likelihood-ratio function, thus it appears as a x10 multiplication in the log-likelihood functions and the log-likelihood ratio functions.
* Calculating statistic value and p-value:
\begin{align}
\lambda (y) &= 2(80log0.8 + 40log0.4 + 110log1.1 + 90log0.9 + 110log1.1 + 170log1.7) \\
&= 94.4
\\
\chi^2_{5, 1-\text{p-value}} &\stackrel{!}{=} 94.4 \\

* Again critical Chi-Squared values remain the same:

And this is no surprise since both significance levels are much greater than the p-value. **This is a further no surprise**, because we expected the Chi-Squared-Test and the Log-Likelihood-Test to be consistent!

**To conclude**, it appears that for the same frequency of occurrences after a bigger number of repeating experiments $N$, insignificant evidence as in section 2 can become highly significant as in section 3. This is because (analogously to the Law of Large Numbers) a relatively unlikely event becomes even more relatively unlikely as the experiment is done repeatedly.


```python
# imports for the tutorial
import numpy as np
import pandas as pd
from scipy.stats import t as t_dist
from scipy.stats import chi2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# chi-squared test
observed = [8, 4, 11, 9, 11, 17]
num_categories = len(observed)
dof = num_categories - 1 # degrees of freedom
total = sum(observed)
expected = [total/num_categories] * num_categories
print(f"Observed: {observed}")
print(f"Expected: {expected}")
alphas = [0.99, 0.95, 0.05, 0.01]

# compute chi-squared statistic
chi_square_stat = np.sum((np.array(observed) - np.array(expected))**2 / np.array(expected))
print("chi-squared statistic value: ", chi_square_stat)

# compute chi-squared p-value
chi_square_p_val = 1 - chi2.cdf(chi_square_stat, df=dof)
print("chi-squared p-value: ", chi_square_p_val)

# compute critical value using the inverse of the CDF
for alpha in alphas:
    critical_value = chi2.ppf(q=1-alpha, df=dof)
    print(f"For alpha={alpha}, critical value: {critical_value}")
    if chi_square_stat > critical_value:
        print("H_0 is rejected!")
    else:
        print("H_0 is accepted!")
    
    # plot rejection/acceptance regions
    x = np.linspace(0, 50)
    fig = plt.figure(figsize=(8, 5))
    ax = fig.add_subplot(111)
    ax.plot(x, chi2.pdf(x, df=dof))

    accept_region_xs = np.arange(0, critical_value, 0.001)
    ax.fill_between(accept_region_xs, chi2.pdf(accept_region_xs, df=dof), color='green')
    reject_region_xs = np.arange(critical_value, 50, 0.001)
    ax.fill_between(reject_region_xs, chi2.pdf(reject_region_xs, df=dof), color='red')
    ax.grid()
    ax.set_title(f"Chi-Squared Acceptance/Rejection Regions for alpha = {alpha}")

# Log-likelihood
observed = np.array(observed)
print(2*(observed.T @ np.log(6*observed/total)))
```

#### <img src="https://img.icons8.com/dusk/64/000000/idea.png" style="height:30px;display:inline"> Solution 2 - Advertisements - To Believe or Not To Believe?
---
1. If we define the i.i.d RVs $X_i \sim Bernoulli(p)$ as the response of the $i^{th}$ dentist (1=recommended), then the mean $\mathbb{E}[X]=\mu$ equals $p$ the rate of recommenders in the population. Now the hypotheses may relate to the mean $\mu=p$.<br>
The question explicitly stated "more than", suggesting composite one sided null & alternative hypotheses, $ \{ H_0: p > 0.75 ; H_1: p \leq 0.75 \} $. However, we shall define a simpler version of the null & alternative hypotheses, whose solution is more reasonable:
\begin{align}
&H_0: p = 0.75 \\
&H_1: p < 0.75
\end{align}

2. A test is well defined by its test-statistic $T$ and its rejection-region $R$. We will build a test of size $\alpha$ as suggested:
\begin{cases} 
T(x) := \Lambda(x) = \frac{f_{\theta_1}(x)}{f_{\theta_0}(x)} \\
R := \{x: \Lambda(x) > k_\alpha\} \\
\end{cases}
Thanks to Neyman-Pearson Lemma, we know that a test of size $\alpha$ of the following form is optimal in a UMP manner:
\begin{cases} 
T^*(x) := \Lambda^*(x) = \frac{f_{\theta_0}(x)}{f_{\theta_1}(x)} \\
R^* := \{x: \Lambda^*(x) < d_\alpha\} \\
\end{cases}
We shall now take the critical value to be $ k_\alpha := \frac{1}{d_\alpha} $ and show that the two rejection-regions are the same.
\begin{align}
R =  \{x: \Lambda(x) > k_\alpha\} = \{x: \frac{f_{\theta_1}(x)}{f_{\theta_0}(x)} > \frac{1}{d_\alpha}\} = \{x: \frac{f_{\theta_0}(x)}{f_{\theta_1}(x)} < \frac{d_\alpha}{1}\} = \{x: \Lambda^*(x) < d_\alpha\} = R^*
\end{align}
We have shown that the two tests are equivalent. Thus they have the same size $\alpha$ and the same power function $\beta(\theta)$, and since the Nayman-Pearson test is a Uniformly Most Powerful test, so is our test.
* **Note:** Since our hypotheses are composite, we were forced to use the Generalized-Nayman-Pearson test, which uses supremum. However, with the following slight extension, the proof will remain the same: $$ f_{\theta_i}(x) \leftarrow \sup_{\theta \in \Theta_i} f_{\theta_i}(x) $$

3. Let the size of the test be $\alpha$, that is the probability of a False-Positive (type-I) error. Now the new test is:
\begin{cases} 
T(x) := \overline{x}_n = \frac{1}{N}\sum_{i=1}^N x_i \\
R := \{x: \overline{x}_n < C\} \\
\end{cases}
Recall $ X \sim Bernoulli(p) \Longrightarrow Var[x] = p(1-p) $, and we can (approximately) use the Z-test Statistic to find critical values, using the CLT.
$$ \overline{X}_n \sim \mathcal{N}(p,\frac{p(1-p)}{n}) $$
$$ z(x) := \frac{\overline{X}_n - \mu}{ \frac{\sigma}{\sqrt{n}}} = \sqrt{\frac{n}{p(1-p)}} (\overline{X}_n - p)  \sim \mathcal{N}(0,1) $$
Now we need to find $C=C(\alpha)$:
\begin{align}
\alpha = Pr\{\text{False-Positive}\}
&= P_{H_0}(\overline{X}_n < C) \\
&= P_{H_0}\Big(z(X) < \sqrt{\frac{n}{p(1-p)}} (C - p)\Big) \\
&= P\Big(z(X) < \sqrt{\frac{390}{0.75(1-0.75)}} (C - 0.75)\Big) \\
&= \Phi(45.61(C - 0.75)) \\
\end{align}
And now $C$ can be calculated by:
$$ C = 0.0219\cdot\Phi^{-1}(\alpha) + 0.75 $$

4. For $\alpha=0.025$ we get $ C = 0.0219 \cdot (-1.96) + 0.75 = 0.707 $. Since $ \overline{x} = 273 / 390 = 0.7 < 0.707 $ we reject the $H_0$ !
5. Now it is given that the null hypothesis is false. This means that $p < 0.75$.<br>
We are asked to calculate $ Pr\{\text{False-Negative}\} = \beta(p)$ for $p_1 = 0.74$ and $p_2 = 0.7$.
\begin{align}
\beta(p) = Pr\{\text{False-Negative}\}
&= P_{H_1}(\overline{X}_n \geq C) \\
&= P_{H_1}\Big(z(X) \geq \sqrt{\frac{n}{p(1-p)}} (C - p)\Big) \\
\end{align}
\begin{align}
\beta(p_1) &= P\Big(z(X) \geq \sqrt{\frac{390}{0.74(1-0.74)}} (0.707 - 0.74)\Big) \\
&= 1 - \Phi(-1.486) \\
&= 1 - 0.071 \\
&= 0.929 \\
\end{align}
\begin{align}
\beta(p_2) &= P\Big(z(X) \geq \sqrt{\frac{390}{0.7(1-0.7)}} (0.707 - 0.7)\Big) \\
&= 1 - \Phi(+0.302) \\
&= 1 - 0.626 \\
&= 0.374 \\
\end{align}
The type-II errors turned out to be quite high, however this does make sense:
* The calculation of the probability of an error does not take into account the realization $\overline{x}_n=0.7$, so we can ignore that value completely.
* A significance level of $\alpha=0.025$ for a one sided test requires quite a strong evidence against $H_0$, and it is hard to get. This is why most of the time we will tend to accept $H_0$, sometimes even though it is a false hypothesis, and that is why a type-II error (False-Negative) is likely.
* Moreover, the rejection of $H_0$ in such a strong test requires a very distinct empirical mean from $p_0=0.75$, whose probability of occurrence  increase as the true mean population of $p_a$ becomes more distinct from the false one.
* $p_a=p_1=0.74$ is simply not distinct enough to produce a sufficiently distinct empirical mean in reasonable probablity.
* $p_a=p_2=0.7$ is distinct enough to produce a sufficiently distinct empirical mean in reasonable probablity, however not that high.
* $p_a=C=0.707$ is the exact point of 0.5:0.5, because the probability of a super-threshold empirical mean and probability of a sub-threshold empirical mean are symmetric at that point.


## <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 3 - t-SNE Gradient

Recall the objective of t-SNE algorithm: $$ C=KL(P||Q) =  \sum_{k,l\neq k}p_{lk}\log\frac{p_{lk}}{q_{lk}} $$
Calculate the gradient of the objective function with respect to the low-dimensionality mappings $y_i$, i.e. calculate $\frac{\partial C}{\partial y_i}$.

Hint:
   * Denote $(1 + ||y_i-y_j||^2)^{-1} = E_{ij}^{-1}$.
       * Notice that $E_{ij} = E_{ji}$.

#### <img src="https://img.icons8.com/dusk/64/000000/idea.png" style="height:30px;display:inline"> Solution 3 - t-SNE Gradient
---
Recall the elements $p_{ij}$-s of $P$:
\begin{align}
p_{ij} &= \frac{e^{-|| x_i - x_j||^2 / 2\sigma^2}}{\sum_k\sum_{l \neq k}e^{-||x_k-x_l||^2 / 2\sigma^2}} \\
&\text{(!) Note that all }p_{ij}\text{-s are independent of all } y_m \text{-s and symmetric: }p_{ij} = p_{ji} \\
\\
\end{align}

Recall the elements $q_{ij}$-s of $Q$:
\begin{align}
q_{ij} &= \frac{(1 + || y_i-y_j ||^2)^{-1}}{\sum_k \sum_{l \neq k}(1 + || y_k - y_l ||^2)^{-1}} \\
&(!)\text{ Note that all }q_{ij}\text{-s are symmetric: }q_{ij} = q_{ji} \\
\\
\end{align}

We denote the following:
\begin{align}
\forall i,j: E_{ij} &:= (1 + || y_i-y_j ||^2) \\
&(!)\text{ Note that all }E_{ij}\text{-s are symmetric: }E_{ij} = E_{ji} \\
Z &:= \sum_k \sum_{l \neq k}E_{kl}^{-1} = \sum_k \sum_{l \neq k}(1 + || y_k - y_l ||^2)^{-1} \\
&(!)\text{ Note that }Z\text{ is the same for all }(i,j) \\
\Longrightarrow q_{ij} &= \frac{E_{ij}^{-1}}{Z} \\
\\
\end{align}

First we expand the objective function:
\begin{align}
C \triangleq KL(P||Q) &= \sum\limits_{k,l\ne k}{{{p}_{lk}}\log \frac{{{p}_{lk}}}{{{q}_{lk}}}} \\ 
 & =\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {{p}_{lk}}}-\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {{q}_{lk}}} \\ 
 & =\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {{p}_{lk}}}+\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {{E}_{lk}}}+\sum\limits_{k,l\ne k}{{{p}_{lk}}\log Z} \\
 &:={{S}_{1}}+{{S}_{2}}+{{S}_{3}} \\
 \\
\end{align}

A few mid-step derivatives:<br>
1)
\begin{align}
  & \frac{\partial {{E}_{lk}}}{\partial {{y}_{i}}}=\frac{\partial }{\partial {{y}_{i}}}\left\{ 1+{{\left\| {{y}_{l}}-{{y}_{k}} \right\|}^{2}} \right\} \\ 
 & l=i,k\ne i\to \frac{\partial {{E}_{ik}}}{\partial {{y}_{i}}}=2({{y}_{i}}-{{y}_{k}}) \\ 
 & l\ne i,k=i\to \frac{\partial {{E}_{li}}}{\partial {{y}_{i}}}=2({{y}_{i}}-{{y}_{l}}) \\ 
 & Otherwise\to \frac{\partial {{E}_{lk}}}{\partial {{y}_{i}}}=0 \\ 
 \\
\end{align}

2)
\begin{align}
  & \frac{\partial \log {{E}_{lk}}}{\partial {{y}_{i}}}={{E}_{lk}}^{-1}\frac{\partial {{E}_{lk}}}{\partial {{y}_{i}}} \\ 
 & l=i,k\ne i\to \frac{\partial {{E}_{ik}}}{\partial {{y}_{i}}}=2({{y}_{i}}-{{y}_{k}}){{E}_{ik}}^{-1} \\ 
 & l\ne i,k=i\to \frac{\partial {{E}_{li}}}{\partial {{y}_{i}}}=2({{y}_{i}}-{{y}_{l}}){{E}_{li}}^{-1} \\ 
 & Otherwise\to \frac{\partial {{E}_{lk}}}{\partial {{y}_{i}}}=0 \\ 
 \\
\end{align}

3)
\begin{align}
  & \frac{\partial ({{E}_{lk}}^{-1})}{\partial {{y}_{i}}}=-{{E}_{lk}}^{-2}\frac{\partial {{E}_{lk}}}{\partial {{y}_{i}}} \\ 
 & l=i,k\ne i\to \frac{\partial {{E}_{ik}}}{\partial {{y}_{i}}}=-2({{y}_{i}}-{{y}_{k}}){{E}_{ik}}^{-2} \\ 
 & l\ne i,k=i\to \frac{\partial {{E}_{li}}}{\partial {{y}_{i}}}=-2({{y}_{i}}-{{y}_{l}}){{E}_{li}}^{-2} \\ 
 & Otherwise\to \frac{\partial {{E}_{lk}}}{\partial {{y}_{i}}}=0 \\ 
 \\
\end{align}

4)
\begin{align}
  \frac{\partial Z}{\partial {{y}_{i}}}&=\frac{\partial }{\partial {{y}_{i}}}\sum\limits_{k}{\sum\limits_{l\ne k}{E_{kl}^{-1}}} \\ 
 & =\sum\limits_{k}{\sum\limits_{l\ne k}{\frac{\partial }{\partial {{y}_{i}}}E_{kl}^{-1}}} \\ 
 & =\sum\limits_{m}{\left\{ (-2({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-2})+(-2({{y}_{i}}-{{y}_{m}}){{E}_{mi}}^{-2}) \right\}} \\ 
 & =-4\sum\limits_{m}{({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-2}} \\ 
 &\text{with the next to last step using the trick we have seen in class} \\
 &\text{(of zeroed elements) and the last step from symmetry} \\
\\
\end{align}

5)
$$ \frac{\partial \log Z}{\partial {{y}_{i}}}={{Z}^{-1}}\frac{\partial Z}{\partial {{y}_{i}}}=-4{{Z}^{-1}}\sum\limits_{m}{({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-2}} $$


We can derive each sum separately with respect to $y_i$:
\begin{align}
\frac{\partial S_1}{\partial y_i} &= \frac{\partial }{\partial y_i}\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {{p}_{lk}}} \equiv 0 \\
&\text{because all }p_{ij}\text{-s are independent of }y_i \\
\\
\frac{\partial S_2}{\partial y_i} &= \frac{\partial }{\partial y_i}\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {{E}_{lk}}} \\
&= \sum\limits_{k,l\ne k}{{{p}_{lk}}\frac{\partial }{\partial y_i}\log {{E}_{lk}}} \\
&= \sum\limits_{m}{{{p}_{im}} \cdot 2({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-1} + {{{p}_{mi}} \cdot 2({{y}_{i}}-{{y}_{m}}){{E}_{mi}}^{-1}}} \\
&= 4\sum\limits_{m}{{{p}_{im}}({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-1}}\\
 &\text{with the next to last step using the trick we have seen in class} \\
 &\text{(of zeroed elements) and the last step from symmetry} \\
\\
\frac{\partial S_3}{\partial y_i} &= \frac{\partial }{\partial y_i}\sum\limits_{k,l\ne k}{{{p}_{lk}}\log {Z}} \\
&= \frac{\partial }{\partial y_i} \log {Z} \sum\limits_{k,l\ne k}{{{p}_{lk}}} \\
&= \frac{\partial }{\partial y_i}\log {Z} \\
&= -4{{Z}^{-1}}\sum\limits_{m}{({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-2}} \\
 &\text{with the second step since }Z\text{ is independent of }k,l \\
 &\text{and the third step because a sum of a whole probability equals 1} \\
\\
\end{align}

To wrap it up:
\begin{align}
\frac{\partial C}{\partial {{y}_{i}}}&=\frac{\partial {{S}_{1}}}{\partial {{y}_{i}}}+\frac{\partial {{S}_{2}}}{\partial {{y}_{i}}}+\frac{\partial {{S}_{3}}}{\partial {{y}_{i}}} \\ 
 & =0+4\sum\limits_{m}{{{p}_{im}}({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-1}}-4{{Z}^{-1}}\sum\limits_{m}{({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-2}} \\ 
 & =4\sum\limits_{m}{\left\{ {{p}_{im}}({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-1}-{{E}_{im}}^{-1}{{Z}^{-1}}({{y}_{i}}-{{y}_{m}}){{E}_{im}}^{-1} \right\}} \\ 
 & =4\sum\limits_{j}{({{p}_{ij}}-\frac{{{E}_{ij}}^{-1}}{Z}){{E}_{ij}}^{-1}({{y}_{i}}-{{y}_{j}})} \\ 
 & =4\sum\limits_{j}{({{p}_{ij}}-{{q}_{ij}}){{E}_{ij}}^{-1}({{y}_{i}}-{{y}_{j}})} \\ 
 & =4\sum\limits_{j}{({{p}_{ij}}-{{q}_{ij}}){{(1+||{{y}_{i}}-{{y}_{j}}|{{|}^{2}})}^{-1}}({{y}_{i}}-{{y}_{j}})} \\ 
\end{align}

1. First we observe the optimization problem and break it down.
    * The problem includes a max iteration on vectors $a \in \mathbb{R}^{D_x}, b \in \mathbb{R}^{D_y}$
        $\mathbb{E}\left[(a^Tx)(b^Ty)\right] = \mathbb{3}\left[(a^Tx)(b^Ty)^T\right] = \mathbb{E}\left[a^Txy^Tb\right]=$
        $= a^t \text{cov(x,y))}b \in \mathbb{R}^{D_x\times D_y} = a^TU\Sigma Vb$
        * a,b are determinant
        * SVD breakdown
        We can cleanly see that max can be achieved (regarding constraints) with vector a being an orthogonal vector in U
        and b a vector in V that corresponds to the highest value in $\Sigma$ which is almost diagonal.
      Furthermore we have 2 constraints $\mathbb{E}\left[(a^Tx)^2\right]=1, \mathbb{E}\left[(b^Ty)^2\right]=1 $
      acting as regularization for both vectors.
        $\mathbb{E}\left[(a^Tx)^2\right]=\mathbb{E}\left[a^Txx^Ta\right]=a^TP_xa=1$
        It is used to keep a from blowing up 
   * This problem is a generalized PCA problem, as we saw in the lectures PCA can be described as an SVD problem.
     Here we solve it the same way, however we don't corralate a RV to itself, but to a diffrent RV.
     if x=y then we have the PCA problem.
2. The problem similar to PCA can be viewed from a geometric or a statistical POV.
    The geometric view tries to get as close a fit to a vector to the data clusters with a MSE approach.
    On the other hand a statistical view approach the data as eliptical clusters and tries to exptact the highest variance
    out of it. Both ways will get us the same equations eventually. Minimazing the error will result falling to the 
    main vector of the a cluster. MSE will include a constant subtructed by the representaion which will result
    in the given max problem.
3. Kernel PCA differs from regular PCa by instead of getting a covariance matrix $\mathbb{R}^{D\times D}$
   We get an $N\times N$ matrix.
   Considering $X\in \mathbb{R}^{d\times n}$, we will note that both matricies are PSD and can be broken down
   
   $XX^T=V\Sigma V^T \text{s.t} V\in \mathbb{R}^{d\times d}$
   
   $X^TX=U\Lambda U^T \text{s.t} U\in \mathbb{R}^{n\times n}$
   
   Lets take the matrix multiplication $(XX^T)(XX^T) $
   
   $(XX^T)(XX^T)=V\Sigma V^TV\Sigma V^T= V\Sigma^{0.5}\Sigma\Sigma^{0.5}V^T$
   
   $X(X^TX)X^T=XU\Lambda U^TX^T=XU\Lambda (XU)^T$
   
   From singularity we can deduce
   
   $V\Sigma^{0.5} = XU$
   
   $\Sigma = \Lambda$
   
   Considering K is the kernel $K=X^TX$ and if x is mean 0 there is no need to bias JKJ = K
   In the lectures we learnt that PCA and KPCA are represented like so
   
   PCA:  $Z=V_k^TX$
   
   KPCA: $Z=\Lambda^{-0.5}U_k^TX^TX = \Lambda^{-0.5}(XU_k)^TX = \Lambda^{-0.5}(V_k\Sigma^{0.5})^TX = $
   
   $ = \Lambda^{-0.5}\Sigma^{0.5}V_k^TX = V_K^TX$
   
   Which means PCA and KPCA with linear kernel gets the same result

```python
# imports
# you can add more if you need
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_s_curve , make_circles
from sklearn.decomposition import PCA , KernelPCA
%matplotlib inline
```

```python
def KPCA(X, d, kernel):

    n = X.shape[1] # Get the number of x vectors
    K = kernel(X, X) # Calculate kernel
    J = np.diag(np.ones(n)) - (np.ones([n, n]) / n) 
    K = J @ K @ J
    print(r'Kernel Shape is {}'.format(K.shape))
    
    eig_val , eig_vec = np.linalg.eig(K) # Break down to eigenvectors and eigenvalue
    index = np.argsort(-eig_val)[:d] # Indecies of the largest eigen values
    print(r'Min eig val is {}, max eig val is{}'.format(min(eig_val), max(eig_val)))
    eig_vec = eig_vec[:, index] # dim{n,k}
    eig_val = eig_val[index] ** -0.5 # Lambda ^-0.5
    print(r'Eigen values are {}'.format(eig_val))
    print(r'Eigen vectors are {}'.format(eig_vec.shape))
    return np.diag(eig_val) @ eig_vec.T @ K # GEt representation
```

```python
kernel_linear = lambda x, y : x.T @ y # Plain linear kernel
kernel_pol = lambda x, y : (1 + x.T @ y) ** 2 # second order polonomial
```

```python
def ker_gau(x, y, s): # Gauusing kernel
    n = x.shape[1]
    ker = np.zeros([n, n])
    for i in range(n):
        for j in range(n):
            v = x[:,i] - y[:,j] # subtruct all vectors from one to other
            ker[i,j] = np.exp(- (v @ v) / (2*s**2))
    return ker

kernel_gauss = lambda x ,y : ker_gau(x, y, 0.5)
```

```python
X, y = make_circles(n_samples=100, random_state=123, noise=0.1, factor=0.2)
print(X.shape)
X_kpca = KPCA(X.T, 2, kernel_gauss).T # Calculate KPCA
# plot
fig = plt.figure(figsize=(10,4))
# original data
ax1 = fig.add_subplot(1,2,1)
ax1.scatter(X[y==0, 0], X[y==0, 1], color='red', marker='^', alpha=0.5)
ax1.scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o', alpha=0.5)
ax1.grid()
ax1.set_title("original data")
# pca
ax2 = fig.add_subplot(1,2,2)
ax2.scatter(X_kpca[y==0, 0], X_kpca[y==0, 1], color='red', marker='^', alpha=0.5)
ax2.scatter(X_kpca[y==1, 0], X_kpca[y==1, 1], color='blue', marker='o', alpha=0.5)
ax2.grid()
ax2.set_title("KPCA")
ax2.set_xlabel("PC1")
ax2.set_ylabel("PC2")
```
