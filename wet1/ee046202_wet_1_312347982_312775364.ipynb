We don't actually care about the counties, and thus we will perform the following preprocessing steps:

* For each state, first change percentage fields (*_pct*) to numbers, and then *sum* for each state and finally re-calculate the percentage out of the new total population. Note that the percentage is out of the *total_population*.
* For each state, calculate the *weighted median* of the the household median and then calculate the mean (average of medians)
    * We want the household median for each state, and in order to that we need to compute the weighted median of each county ($median * \text{total_population}$) and then the best we can do (as we don't that data available) is calculate the mean.

**Hints**: `pandas` has great tools for doing all of this. Check out: `df[df['state']=='Alabama']`, `df.uniuqe()`, `df.sum()`, `df.append()`...

```python
df_clean = df.dropna(axis=0, how='any') # Remoce all counties with at least 1 NaN field
count = ["trump16", "clinton16", "otherpres16", "total_population", "cvap"] # countable fields
percent = [col for col in df.columns if "_pct" in col] # percentagew fields
db = [] # new data base
for st in df["state"].unique(): # go through each state
    df_ = df_clean.loc[df["state"]==st] # get just the counties in the state
    d = {"state": st} # dictionary for each state
    for col in count:
        d[col] = df_[col].sum() # sum all from a county for each countable field
    pop = np.array(df_["total_population"])
    for col in percent:
        pct = np.array(df_[col])
        d[col] = pct.dot(pop) / d["total_population"] # calculate weighted pct mean
    med = np.array(df_["median_hh_inc"])
    d["median_hh_inc"] = med.dot(pop) / d["total_population"] # calculate weighted mean
    d["trump16_pct"] = 100 * d["trump16"] / (d["trump16"] + d["clinton16"] + d["otherpres16"]) 
    db.append(d)
prep_df = pd.DataFrame(db)
prep_df
```

### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 2 - Point Estimates
---
As learned in class, point estimates are estimates of population parameters based on sample data. For instance, if we wanted to know the average age of registered voters in the U.S., we could take a survey of registered voters and then use the average age of the respondents as a point estimate of the average age of the population as a whole. The average of a sample is known as the sample mean. The sample mean is usually not exactly the same as the population mean. This difference can be caused by many factors including poor survey design, biased sampling methods and the randomness inherent to drawing a sample from a population

In this task, we will go back to the original data, `df` in the code, and we will treat each *county* as one vote and we will try to estimate the **mean** *median income* of voters.

* Constant the seed (`np.random.seed(0)`)
* Clean the original data by removing counties with NaN (as before, use `df.dropna()`)
* Calculate the **true** mean of the *median income* (print it)
* Take a random sample of $n=1000$ samples, calculate the sample mean, and calculate the difference from the true mean (print them)
* Repeat the process of sampling $n=1000$ (with replacement) $N=10, 20, 40, 50, 100, 200, 500, 1000, 2000$ times, and for each $N$ calculate the mean, calculate the difference from the true mean, plot the difference from the true mean vs. $N$.

```python
df_clean = df.dropna(axis=0, how='any')
cnt_inc = np.array(df_clean["median_hh_inc"])

mm_hh_inc = np.mean(cnt_inc) # We consider each county as a voters so no need for a weighted mean
print("Mean of the median house hold income is {:.2f}$".format(float(mm_hh_inc)))

np.random.seed(0)
n = 1000
N_lst = [1, 10, 20, 40, 50, 100, 200, 500, 1000, 2000]
res = []
for N in N_lst:
    mm_rnd = []
    for i in range(N):
        idx = np.random.choice(range(cnt_inc.size), n) # Choose random indecies
        mm_rnd.append(np.mean(cnt_inc[idx]))
    res.append(np.mean(mm_rnd))
diff = res - mm_hh_inc
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.plot(N_lst, diff)
ax.grid()
ax.set_title("Sample estimation , sample size = 1000")
ax.set_xlabel("N")
ax.set_ylabel("Bias")
```

### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 3 - Confidence Intervals
---
A point estimate can give you a rough idea of a population parameter like the mean, but estimates are prone to error and taking multiple samples to get improved estimates may not be feasible (would you call 500 people, 2000 times?). A confidence interval is a range of values above and below a point estimate that captures the true population parameter at some predetermined confidence level. For example, if you want to have a 95% chance of capturing the true population parameter with a point estimate and a corresponding confidence interval, you'd set your confidence level to 95%. Higher confidence levels result in a wider confidence intervals.

Calculating a confidence interval -  taking a point estimate and then adding and subtracting a margin of error to create a range. Margin of error is based on your desired confidence level, the spread of the data and the size of your sample. The way you calculate the margin of error depends on whether you know the standard deviation of the population or not.

* Calculate a 90%, 95% and a 99% confidence interval for the **mean** *median income* of voters. Take a sample of $n=1000$.
    * Equivalently, $\alpha=[0.1, 0.05, 0.01]$
    * Don't forget to **standartize** the data. and notice that $\sigma \neq 1$ as in the tutorial.
    * Use the true STD of the population
    * To calculate the inverse of the CDF, use `scipy.stats.norm.ppf()`
* Compare your answer to using `scipy.stats.norm.interval(alpha=, loc=, scale=)`. Read the doc, use Shift + Tab (only in Jupyter) inside the parenthesis to understand how to use it.
    * You need to print the result for each $\alpha$, with both methods

```python
np.random.seed(0)

cnt_inc = np.array(df_clean["median_hh_inc"])
sig = np.std(cnt_inc) # STD of the data
n = 1000
idx = np.random.choice(range(cnt_inc.size), n) # Selected random indcies
alpha = [0.1, 0.05, 0.01]
ci_calc = []
ci_true = []
mean = np.mean(cnt_inc[idx]) # Calculated mean
for a in alpha:
    ci = scipy.stats.norm.ppf(1-a/2) # Gets the margin the point of a/2 in CDF
    ci = sig * ci / np.sqrt(n) # Scales the margin to the given distribution 
    ci_calc.append((mean - ci, mean + ci))
    print(r"The calculated confidence interval of {}% the mean midean is {:.0f} - {:.0f} USD".format(100*(1-a), mean-ci, mean+ci))
    low, high = scipy.stats.norm.interval(1-a, loc = mean, scale=sig/np.sqrt(n))
    ci_true.append((low,high))
    print(r"The confidence interval of {}% the mean midean is {:.0f} - {:.0f} USD".format(100*(1-a), low, high))
```

Point estimates and confidence intervals are basic inference tools that form the foundation for another inference technique: statistical hypothesis testing. Statistical hypothesis testing is a framework for determining whether observed data deviates from what is expected. The T-test is a statistical test used to determine whether a numeric data sample differs significantly from the population or whether two samples differ from one another.

In this exercise we will perform the t-test for the mean of unemployment fraction (`clf_unemploy_pct`) in the US. We will use the pre-procesesd data (`prep_df`) and a sample of size $n=10$. We wish to test whether the sample mean differs from the population mean. The null hypothesis, $H_0$ is that the mean is not different. Recall that we are performng t-tests when we are not given the true mean and true STD. The steps:

* Constant the seed
* Calculate the true mean of unemployment fraction (use fractions and not percents: `clf_unemploy_pct / 100`)
    * In t-test, you assume that the mean ($\mu$) is unknown, then you formulate the $H_0: \mu= \mu_0$. So if $\mu$ is unknown, how do you calculate the t-statistic? Well, assume that somebody told you that in the last elections, $\mu = C$, where $C$ is the also the true mean of this election.
    * In other words, to calculate t-statistic, use the true mean. 
* Take a sample of $n=10$ and calculate its mean.
* Calculate the sample's unbiased STD.
* Calculate the t-statistic value.
    * Use the true mean for $\overline{\mu} - \mu_0$
* Calculate $t_{n-1, \frac{\alpha}{2}}$ for $\alpha=[0.1, 0.05, 0.01]$ significance levels (how many derees of freedom are there for the t-distribution?). Use `scipy.stats.t.ppf()`.
* Calculate the p-value (don't forget that this is a **two-sided** test). For each level, do we reject $H_0$ or accept it?. Use `scipy.stats.t.cdf()`.
* Compare your results to `scipy.stats.ttest_1samp()`. This function performs the t-test, and outputs the t-statistic and the p-value.

```python
def gen_distinct_colors(n, sat=1.0, val=1.0):
    """Generate an array of n distinct colors"""
    h_lvls = np.linspace(0, 1, n + 1)
    s_lvls = sat * np.ones((n + 1, 1))
    v_lvls = val * np.ones((n + 1, 1))
    hsv_colors = np.column_stack([h_lvls, s_lvls, v_lvls])
    rgb_colors = mpl.colors.hsv_to_rgb(hsv_colors)

    return rgb_colors[:-1,:]
```

```python
def plot_sk_dim_reduction(X, y, title=None, ax=None, marker=None, label="",
                          classes=None, sat=1.0, val=1.0):
    """Plot Scikit dimensionality reduction output (2D or 3D only)"""
    # Check plot dimensions
    ndim = X.shape[1]
    if ndim == 2:
        proj = None
    elif ndim == 3:
        proj = "3d"
    else:
        raise SystemError("For plotting, output dimensions must be 2 or 3!")
    
    # Prepare color pallete
    if classes is None:
        classes = np.unique(y)
    rgb_colors = gen_distinct_colors(len(classes), sat, val)

    # plot
    if ax is None:
        fig = plt.figure(figsize=(8, 8))
        ax = plt.axes(projection=proj)
    for i, cls in enumerate(classes):
        ax.scatter(*X[(y == cls), :].T, marker=marker, color=rgb_colors[i],
                   label=f'Class {cls}{label}')
    # Final touches
    ax.grid(b=True, which='major')
    ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left", prop={"size": 10})
    ax.set_title(title)
```

```python
# Perform PCA
pca = PCA(n_components=2, random_state=SEED)
X_pca = pca.fit_transform(X)

# Plot
plot_sk_dim_reduction(X_pca, y, title="2D PCA of the non-normalized Wine Dataset")
```

```python
# Scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2, random_state=SEED)
X_scaled_pca = pca.fit_transform(X_scaled)

# Plot
plot_sk_dim_reduction(X_scaled_pca, y, title="2D PCA of the Standardized Wine Dataset")
```

#### <img src="https://img.icons8.com/dusk/64/000000/idea.png" style="height:30px;display:inline"> Answer 6 - Importance of Feature Scaling
---
- Usually dimensionality reduction algorithms try to extract the **information content** of every feature (or combination of features), and not the scale itself, because most of the time we would like to keep the data **informative** (with respect to the problem we are trying to solve), while compressing in order to enhance speed, memory and model simplicity.<br>
In fact, it is quite the opposite: the scale does not necessarily have anything to do with information content - i.e. data of hair length theoretically has the same information content when given in micrometers, millimeters or meters.<br>
On the other hand, when these algorithms try to estimate the information within features, they are sensitive to scale nonetheless. PCA, for example, picks the features' linear combinations of highest variance, relying on the strong correlation between variance and information content. PCA is even extra-sensitive to scale because it deals with the squared data (variance, Euclidean distances...).
<br>
Standardization is important in two layers:<br>
1) Centering - even low variance data points can have misleading magnitude when uncentered, i.e. wavelengths of photons from a narrow laser source in nanometers.<br>
2) Normalization - even centered data points can have misleading magnitude when unnormalized, i.e. delta-wavelengths of photons from a wide LED source when given in micrometers.

* In classification problems, we connect the separability between classes to the preserving of information, because after our dimensionality reduction what we see in the plots is all we have.
And indeed, in task 6 it seems that PCA has managed to preserve much more information when executed on the standardized data, as the classes there have kept their clustery and interclass-distinct form.

### <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 7 - T-SNE
---
- Q: Explain in short how t-SNE works and how it is different from PCA.

A:<br>
The idea behind t-SNE, similarly to PCA, is to try to reduce the dimensions of the data while preserving "distances" between points in the higher dimension when transformed into the lower dimension.
<br><br>
On the one hand, PCA's distance definition is the Euclidean distance $L_2$, and it reduces dimensionality by linear transformations only.
<br><br>
On the other hand, t-SNE's distance definition is the KL-Divergence between probabilities $D_{KL}(P||Q)$, and it solves a minimization problem using an iterative algorithm of a random initial guess Momentum Gradient Descent as the iterative step:<br>
What "probabilities"? A probability score $p_{ij}$ is given to every two points in the higher dimension, based on their Euclidean distance there, using Gaussian kernel. A slightly different probability score $q_{ij}$ is given to those same two points in the lower dimensions, based on their Euclidean distance there, using Cauchy kernel = $StudentT(\nu=1)$. Now the algorithm tries to bring $P$ and $Q$ closer together by minimizing $D_{KL}(P||Q)$ as the cost function.<br>
How is it different from PCA?<br>
1) For every two points, the probability score of t-SNE takes into account the distances to the other points as well! So the distance is not a simple, absolute distance as in PCA.<br>
2) The KL-Divergence is not symmetric! While closer points in the higher dimension must be closer still in the lower dimensions (otherwise a big cost is achieved), further points do not necessarily have to be far, because their influence on the cost function are minor. This forces the optimization to focus on closer distances, and by that to **retain local structures** of the data in a different manner than the global structures of the data. PCA prefers to handle global structures (of distant points) first.<br>
3) The use of a heavy-tailed probability kernel, the Student-t one, helps to prevent the Crowding problem, where closer points in the higher dimension data tend to become much closer in the lower dimension data, in order to retain mid-range distances too. The heavy tailed probability encourages points of the lower dimensions data to spread more in order to mimic thier probability score in the higher dimensions data.
This helps the algorithm to **retain global structures** of the data as well!
4) Dimension reduction is achieved by an iterative MGD method, instead of a linear one. As a results, t-SNE is not constrained to retain linear relations alone, unlike PCA.<br><br>


- Q: Compare the results of PCA and t-SNE on the Wine dataset

While the PCA algorithms has seemed to perform not bad on the Wine dataset, the t-SNE has managed to achieve a much better clustering and separation of the data. Furthermore, the existence of three clusters is more visible in the t-SNE's results than in the PCA's (although even then not all points would have been classified correctly, since the groundtruth of each point is an extra info we usually do not have). Thus, the odds for one to mistakenly conclude that the data is consisted of one or two classes are higher for the PCA's results.<br>

```python
def classify_sk_perceptron(X, y):
    """Classify data using a multi-class Perceptron"""
    # Split to train and test sets
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=0.2, random_state=SEED, stratify=y)

    # Initialize a Perceptron classifier
    clf = Perceptron(tol=None, max_iter=10000, random_state=SEED)

    # Train on train-set
    clf.fit(X_train, y_train)

    # Test on test-set
    score = clf.score(X_test, y_test)

    return clf, score
```

```python
def plot_decision_bounds(clf, ax=None):
    """Plot decision bounds for a classifier"""
    # Prepare color pallete
    rgb_colors = gen_distinct_colors(len(clf.classes_))

    # plot
    if ax is None:
        fig = plt.figure(figsize=(8, 8))
        ax = plt.axes()

    # Calculate and plot decision boundaries
    (xmin, xmax), (ymin, ymax) = ax.get_xlim(), ax.get_ylim()
    W = clf.coef_
    b = clf.intercept_
    for i in range(len(clf.classes_)):
        xx = np.linspace(xmin, xmax, 2)
        yy = (-W[i, 0] / W[i, 1]) * xx + (-b[i] / W[i, 1])
        ax.plot(xx, yy, color=rgb_colors[i], linewidth=1)
    ax.set_xlim([xmin, xmax])
    ax.set_ylim([ymin, ymax])
```

```python
def plot_classification_graph(X, y, clf, title="", ax=None):
    """Plot classification graphdecision (data and decision boundaries)"""
    # Restore the split to train and test sets
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=0.2, random_state=SEED, stratify=y)

    # Calculate classification accuracy
    score = clf.score(X_test, y_test)
    
    # plot
    if ax is None:
        fig = plt.figure(figsize=(8, 8))
        ax = plt.axes()
    title += f"\nClassification accuracy: {score:.2%}"
    # Plot correctly classified training points
    mask = (clf.predict(X_train) == y_train)
    plot_sk_dim_reduction(X_train[mask, :], y_train[mask], ax=ax, title=title,
                          marker=".", classes=clf.classes_, sat=0.25,
                          label=" - Train; Correct")
    # Plot incorrectly classified training points
    mask = np.invert(mask)
    plot_sk_dim_reduction(X_train[mask, :], y_train[mask], ax=ax, title=title,
                          marker="x", classes=clf.classes_, sat=0.25,
                          label=" - Train; Incorrect")
    # Plot correctly classified test points
    mask = (clf.predict(X_test) == y_test)
    plot_sk_dim_reduction(X_test[mask, :], y_test[mask], ax=ax, title=title,
                          marker=".", classes=clf.classes_, sat=1.0,
                          label=" - Test; Correct")
    # Plot incorrectly classified test points
    mask = np.invert(mask)
    plot_sk_dim_reduction(X_test[mask, :], y_test[mask], ax=ax, title=title,
                          marker="x", classes=clf.classes_, sat=1.0,
                          label=" - Test; Incorrect")
    # Plot decision boundaries
    plot_decision_bounds(clf, ax)
```

```python
# Classify PCA's 2D output data
clf, score = classify_sk_perceptron(X_scaled_pca, y)
print(f"Classification accuracy for PCA on standardized data: {score:.2%}")
# Plot data and decision boundaries
plot_classification_graph(X_scaled_pca, y, clf,
    title=f"PCA results - Perceptron decision boundaries")

# Classify t-SNE's 2D output data
clf, score = classify_sk_perceptron(X_scaled_tsne, y)
print(f"Classification accuracy for t-SNE on standardized data: {score:.2%}")
# Plot data and decision boundaries
plot_classification_graph(X_scaled_tsne, y, clf,
    title=f"t-SNE results - Perceptron decision boundaries")
```

- Q: Describe and explain what is Perceptron doing.<br>

A:<br>
A Perceptron is an algorithm for a *binary linear classifier*: it learns and then predicts the class of a data point (represented by a feature vector) based on linear combinations of the features alone.<br>
Its decision rule is defined by $ \hat{y} = sign(w^Tx + b) $ for a feature vector $x$, and $w, b$ are tunable weights that adjust in order to achieve the best classification rate on an unseen test data, by training on a given training data.<br>
The perceptron works in epochs. For each sample the weights are updated according to $ w^{(t + 1)} = w^{(t)} + \frac{\eta}{2} \big(y(t)-\hat{y}(t)\big)x(t),$ that adds a non-zero value for bad classification only.<br><br>

- Q: Describe the accuracy results. If there is a difference, explain its source.<br>

A:<br>
The perceptron classification on the PCA's output has achieved better classification rate. It seems that it made the data more linearly separable, specifically when it came to multiclass classification. In our result, the t-SNE output has produced nice, separable clusters, but the perceptron has failed to determine between them because their separation is not linear! This is why the classification of the $C_1$ class, which is located just between the other two, has performed poorly and increased the misclassification rate significantly.<br>
In general, this is not such a big surprise. t-SNE is intended mostly for visualization purposes, ones that humans can understand and respond to. They are not meant for machines. Because of t-SNE's  non-linear behavior, linearly separable data will not necessarily remain linear after the transformation, since linear relations are not necessarily preserved in t-SNE.<br>
PCA, on the other hand, executes a simple linear projection of the higher dimensional data, onto the most significant lower dimensions. Thus, it retains linear relations, and a near linearly separable data such as The Wine Dataset's is more likely to remain linearly separable after the transformation, to some extent. Therefore, it performs very well as a preprocessing algorithm for the linear perceptron-classifier, only when it comes to near linearly separable problems. However, if the input data has been radially separable, for instance, a more sophisticated algorithm should have been used (such as KPCA).

In this task we will compare different dimensionality reduction techniques on the Digits dataset (Each datapoint is a 8x8 image of a digit). Reduce dimensions to `n_components=2`. Plot all the results. You should tune each algorithm's hyper-parameters. Use the table below:

    
|Algorithm| Call to...| Hyper-parameters to tune|
|---------|-----------|-------------------------|
| PCA     | `PCA()`   | None                    |
| KernelPCA | `KernelPCA()`   | `kernel`        |
| t-SNE     |  `TSNE()`       | `perplexity`    |
| Locally Linear Embedding (LLE)|  `LocallyLinearEmbedding()`| `n_neighbors` | 
| Isomap| Isomap() |`n_neighbors` |

```python
# Load Digits dataset
X, y = load_digits(return_X_y=True)
```

```python
# Perform standardization on X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# A dictionary for X_scaled output of all next algorithms
X_scaled_ = {}
```

```python
# Perform PCA on X_scaled
pca = PCA(n_components=2, random_state=SEED)
X_scaled_["PCA"] = pca.fit_transform(X_scaled)
# Plot
plot_sk_dim_reduction(X_scaled_["PCA"], y,
    title="2D PCA on the Digits Dataset")
```

```python
# Perform Kernel PCA on X_scaled
# Tune kernels
ker_array = ["linear", "poly", "rbf", "sigmoid", "cosine"]
# "rbf" (Gaussian) peformed the best! *(seed=666)*
ker_array = ["rbf"]
# Run Kernel PCA test
for ker in ker_array:
    # Perform Kernel PCA
    kpca = KernelPCA(n_components=2, kernel=ker, random_state=SEED)
    X_scaled_["Kernel PCA"] = kpca.fit_transform(X_scaled)

    # Plot
    plot_sk_dim_reduction(X_scaled_["Kernel PCA"], y,
        title=f"2D KernelPCA on the Digits Dataset, kernel: {ker}")
```

```python
# Perform t-SNE on X_scaled
# Tune perplexity
perp_array = range(51)
# Top 11 candidtaes
perp_array = [10, 14, 16, 17, 18, 19, 21, 22, 23, 24, 34, 35]
# 16 peformed the best! *(seed=666)*
perp_array = [16]

# ker_array = ["rbf"]
# Run t-SNE test
for perp in perp_array:
    # Perform t-SNE
    tsne = TSNE(n_components=2, perplexity=perp, random_state=SEED)
    X_scaled_["t-SNE"] = tsne.fit_transform(X_scaled)

    # Plot
    plot_sk_dim_reduction(X_scaled_["t-SNE"], y,
        title=f"2D t-SNE on the Digit Dataset, perplexity={perp}")
```

```python
# Perform LLE on X_scaled
# Tune n_neighbors
nn_array = range(2, 51)
# Final 2 candidates
nn_array = [29, 40]
# 40 peformed the best! *(seed=666)*
nn_array = [40]
# Run LLE test
for nn_lle in nn_array:
    # Perform LLE
    lle = LocallyLinearEmbedding(n_components=2, n_neighbors=nn_lle, random_state=SEED)
    X_scaled_["LLE"] = lle.fit_transform(X_scaled)

    # Plot
    plot_sk_dim_reduction(X_scaled_["LLE"], y,
        title=f"2D LLE on the Digit Dataset, n_neighbors={nn_lle}")
```

```python
# Perform Isomap on X_scaled
# Tune n_neighbors
nn_array = range(3, 51)
# Final 2 candidates
nn_array = [3, 5, 6, 7, 9, 13, 35]
# 3 peformed the best! *(seed=666)*
nn_array = [3]
# Run Isomap test
for nn_isomap in nn_array:
    # Perform Isomap
    isomap = Isomap(n_components=2, n_neighbors=nn_isomap)
    X_scaled_["Isomap"] = isomap.fit_transform(X_scaled)

    # Plot
    plot_sk_dim_reduction(X_scaled_["Isomap"], y,
        title=f"2D Isomap on the Digit Dataset, n_neighbors={nn_isomap}")
```

```python
# Shared plot
algo_params = {"PCA": "",
                "Kernel PCA": f"kernel={ker}",
                "t-SNE": f"perplexity={perp}",
                "LLE": f"n_neighbors{nn_lle}",
                "Isomap": f"n_neighbors={nn_isomap}"}

fig, axes = plt.subplots(nrows=1, ncols=len(algo_params), figsize=(5*len(algo_params), 5))
fig.suptitle("Comparison between 2D dimensionality reduction algorithm on " + 
             "the Digits Dataset", size="xx-large")

# Plot all outputs
for ax, algo in zip(axes, algo_params):
    # Plot
    plot_sk_dim_reduction(X_scaled_[algo], y, ax=ax,
        title=f"2D {algo}; {algo_params[algo]}")
    l = ax.legend()
    l.remove()
ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left", prop={"size": 10});
```
